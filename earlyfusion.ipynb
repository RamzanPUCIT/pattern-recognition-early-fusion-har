Define the LSTM model

```python
# Cell 21: LSTM model
import torch
import torch.nn as nn

class LSTMClassifier(nn.Module):
    def __init__(self, input_dim=27, hidden_dim=64, num_layers=2, num_classes=12, dropout=0.3):
        super(LSTMClassifier, self).__init__()

        self.lstm = nn.LSTM(
            input_size=input_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout
        )

        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(hidden_dim, num_classes)

    def forward(self, x):
        out, _ = self.lstm(x)          # out: (batch, seq_len, hidden_dim)
        out = out[:, -1, :]            # last time step
        out = self.dropout(out)
        out = self.fc(out)
        return out

model = LSTMClassifier(
    input_dim=27,
    hidden_dim=64,
    num_layers=2,
    num_classes=len(class_names),
    dropout=0.3
).to(device)

print(model)
```

Define CNN+LSTM hybrid model

```python
# Cell 29: CNN+LSTM hybrid model
import torch
import torch.nn as nn

class CNNLSTMClassifier(nn.Module):
    def __init__(self, input_dim=27, cnn_channels=64, lstm_hidden=64, num_layers=1, num_classes=12, dropout=0.3):
        super(CNNLSTMClassifier, self).__init__()

        self.conv1 = nn.Conv1d(in_channels=input_dim, out_channels=cnn_channels, kernel_size=3, padding=1)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool1d(kernel_size=2)

        self.lstm = nn.LSTM(
            input_size=cnn_channels,
            hidden_size=lstm_hidden,
            num_layers=num_layers,
            batch_first=True
        )

        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(lstm_hidden, num_classes)

    def forward(self, x):
        # x: (batch, seq_len, features) = (batch, 100, 27)
        x = x.permute(0, 2, 1)              # (batch, 27, 100)
        x = self.conv1(x)                   # (batch, cnn_channels, 100)
        x = self.relu(x)
        x = self.pool(x)                    # (batch, cnn_channels, 50)

        x = x.permute(0, 2, 1)              # (batch, 50, cnn_channels)
        out, _ = self.lstm(x)               # (batch, 50, lstm_hidden)
        out = out[:, -1, :]                 # last time step
        out = self.dropout(out)
        out = self.fc(out)
        return out

cnn_lstm_model = CNNLSTMClassifier(
    input_dim=27,
    cnn_channels=64,
    lstm_hidden=64,
    num_layers=1,
    num_classes=len(class_names),
    dropout=0.3
).to(device)

print(cnn_lstm_model)
```

Subject-independent CNN+LSTM model

```python
# Cell 43: Build CNN+LSTM model for subject-independent setting
import torch
import torch.nn as nn

class CNNLSTMClassifier(nn.Module):
    def __init__(self, input_dim=27, cnn_channels=64, lstm_hidden=64, num_layers=1, num_classes=12, dropout=0.3):
        super(CNNLSTMClassifier, self).__init__()

        self.conv1 = nn.Conv1d(in_channels=input_dim, out_channels=cnn_channels, kernel_size=3, padding=1)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool1d(kernel_size=2)

        self.lstm = nn.LSTM(
            input_size=cnn_channels,
            hidden_size=lstm_hidden,
            num_layers=num_layers,
            batch_first=True
        )

        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(lstm_hidden, num_classes)

    def forward(self, x):
        x = x.permute(0, 2, 1)      # (batch, 27, 100)
        x = self.conv1(x)           # (batch, 64, 100)
        x = self.relu(x)
        x = self.pool(x)            # (batch, 64, 50)

        x = x.permute(0, 2, 1)      # (batch, 50, 64)
        out, _ = self.lstm(x)
        out = out[:, -1, :]
        out = self.dropout(out)
        out = self.fc(out)
        return out

cnn_lstm_si_model = CNNLSTMClassifier(
    input_dim=27,
    cnn_channels=64,
    lstm_hidden=64,
    num_layers=1,
    num_classes=len(class_names),
    dropout=0.3
).to(device)

print(cnn_lstm_si_model)
```
